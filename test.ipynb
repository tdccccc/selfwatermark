```python
def get_jpg_files(folder_path):
    folder_path = Path(folder_path)
    print(folder_path)
    print(os.name)
    if str(os.name) == 'nt':
        jpg_files = list(folder_path.glob('**\\*.jpg'))
        JPG_files = list(folder_path.glob('**\\*.JPG'))
        jpeg_files = list(folder_path.glob('**\\*.jpeg'))
    if str(os.name) == 'posix':
        jpg_files = glob.glob(os.path.join(folder_path, '*.jpg'))
        JPG_files = glob.glob(os.path.join(folder_path, '*.JPG'))
        jpeg_files = glob.glob(os.path.join(folder_path, '*.jpeg'))
    return jpg_files+JPG_files+jpeg_files

# 替换'your_folder_path'为你想要获取文件的文件夹路径
folder_path = 'image/'
jpg_files = get_jpg_files(folder_path)
```

```python
from PIL import Image
from PIL.ExifTags import TAGS, GPSTAGS

def get_exif_data(image_path):
    # 打开图像文件
    image = Image.open(image_path)

    # 获取 EXIF 数据
    exif_data = image._getexif()

    if exif_data is not None:
        # 创建一个字典来存储所需的信息
        exif_info = {}

        # 遍历 EXIF 数据
        for tag, value in exif_data.items():
            tag_name = TAGS.get(tag, tag)
            exif_info[tag_name] = value

        return exif_info
    else:
        print("No EXIF data found in the image.")
```

```python
# 获取 EXIF 数据
exif_info = get_exif_data(jpg_files[2])

# 打印所需信息
if exif_info:
    print("相机品牌和型号:", exif_info.get('Make', 'N/A'))
    print("相机制造商:", exif_info.get('Manufacturer', 'N/A'))
    print("拍摄日期:", exif_info.get('DateTimeOriginal', 'N/A'))
    print("快门速度:", exif_info.get('ExposureTime', 'N/A'))
    print("光圈值:", exif_info.get('FNumber', 'N/A'))
    print("感光度:", exif_info.get('ISOSpeedRatings', 'N/A'))
    print("图像分辨率:", exif_info.get('ImageWidth'), "x", exif_info.get('ImageLength'))
    print("图像方向:", exif_info.get('Orientation', 'N/A'))
    print("焦距:", exif_info.get('FocalLength', 'N/A'))
    print("镜头制造商和型号:", exif_info.get('LensMake', 'N/A'), exif_info.get('LensModel', 'N/A'))
    print("摄影师:", exif_info.get('Artist', 'N/A'))
```
